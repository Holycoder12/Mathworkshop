```python
import numpy as np

def matrix_vector_mult(A, v):
    result = []
    for i in range(len(A)):
        sum = 0
        for j in range(len(v)):
            sum += A[i][j] * v[j]
        result.append(sum)
    return result

A = [[1,2,3],
     [4,5,6],
     [7,8,9]]

v = [1,1,1]

print("Manual:", matrix_vector_mult(A, v))

# NumPy check
A_np = np.array(A)
v_np = np.array(v)

print("NumPy:", A_np.dot(v_np))
```

```python
import numpy as np

def matrix_mult(A, B):
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]

    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]

    return result

A = [[1,2],
     [3,4]]

B = [[5,6],
     [7,8]]

print("Manual:", matrix_mult(A,B))

# NumPy check
A_np = np.array(A)
B_np = np.array(B)

print("NumPy:\n", A_np.dot(B_np))
```

```python
import numpy as np

def is_symmetric(A):
    A = np.array(A)
    return np.array_equal(A, A.T)

A = [[1,2,3],
     [2,5,6],
     [3,6,9]]

print("Is symmetric?", is_symmetric(A))
```
