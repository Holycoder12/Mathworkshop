```python
import numpy as np
import matplotlib.pyplot as plt

# Define x range
x = np.linspace(-5, 5, 200)

# Equations:
# 1) 2x + 3y = 3 → y = (3 - 2x)/3
y1 = (3 - 2*x) / 3

# 2) 4x + 6y = 6 → y = (6 - 4x)/6
y2 = (6 - 4*x) / 6

# 3) 3x - 2y = 1 → y = (3*x - 1)/2
y3 = (3*x - 1) / 2

# Solve intersection of eqn (1) and (3)
A = np.array([[2, 3],
              [3, -2]])
b = np.array([3, 1])
intersection = np.linalg.solve(A, b)

# Plotting
fig, ax = plt.subplots(figsize=(6,6))

ax.plot(x, y1, 'r', label='2x + 3y = 3')
ax.plot(x, y2, 'g', label='4x + 6y = 6')
ax.plot(x, y3, 'b', label='3x - 2y = 1')

# Mark intersection point
ax.scatter(intersection[0], intersection[1], color='black', s=60, zorder=5)
ax.text(intersection[0]+0.2, intersection[1], f"({intersection[0]:.2f}, {intersection[1]:.2f})", color='black')

# Axes setup
ax.axhline(0, color='k', linewidth=1)
ax.axvline(0, color='k', linewidth=1)
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect('equal')
ax.grid(True, ls=':')

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

print("Intersection point of eqn (1) and (3):", intersection)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([-3,-2,-1,0,1,2,3,4,5,7])
y = np.array([-7.6,1,4.8,5,2.8,-0.6,-4,-6.2,-6,6.4])

coeff = np.polyfit(x, y, 3)

print("Polynomial coefficients:", coeff)

p = np.poly1d(coeff)
xp = np.linspace(-3,7,100)

plt.scatter(x,y)
plt.plot(xp, p(xp))
plt.show()
```
