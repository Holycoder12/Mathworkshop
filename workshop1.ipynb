```python
def myPow(x, n):
    result = 1
    count = 0

    while count < n:
        result *= x
        count += 1

    return result


print(myPow(2, 3))
```

```python
def fib(n):
    a, b = 0, 1
    count = 0

    while count < n:
        print(a, end=" ")
        a, b = b, a + b
        count += 1


n = int(input("Enter number of terms you wise: "))
fib(n)
```

```python
def SumDigits(n):
    total = 0

    while n > 0:
        digit = n % 10
        total += digit
        n //= 10

    print(total)


SumDigits(234)
```

```python
def squaredSum(arr):
    total = 0
    i = 0

    while i < len(arr):
        total += arr[i] * arr[i]
        i += 1

    return total


print(squaredSum([5, 3, 4]))
```

```python
def sumDigits(num):
    total = 0

    while num > 0:
        digit = num % 10
        total += digit
        num //= 10

    return total


print(sumDigits(123456))
```
