question 2

```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

Question 3

```python
def find_twin_primes(limit):
    for i in range(3, limit - 1, 2):
        if is_prime(i) and is_prime(i + 2):
            print(f"({i}, {i+2})")

find_twin_primes(1000)
```

Question 4

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total
```

Question 5

```python
def find_next_perfect(start_after):
    n = start_after + 1
    while True:
        if sumOfDivisors(n) == n:
            return n
        n += 1

print(find_next_perfect(6)) # Output: 28
```

Question 6

```python
marks = [20, 30, 35, 55, 70, 80, 90]

# Mean
mean = sum(marks) / len(marks)

# Median
marks.sort()
n = len(marks)
if n % 2 == 0:
    median = (marks[n//2 - 1] + marks[n//2]) / 2
else:
    median = marks[n//2]

print(f"Mean: {mean}, Median: {median}")
```

Question 7

```python
def decimal_to_binary(n):
    if n == 0: return "0"
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //= 2
    return binary
```

Question 8

```python
def prodDigits(n):
    product = 1
    n = abs(n) # Handle negative numbers
    while n > 0:
        product *= (n % 10)
        n //= 10
    return product

print(prodDigits(123)) # Output: 6
```
