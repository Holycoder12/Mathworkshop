```python
import matplotlib.pyplot as plt
import random
x = [random.randint(0, 50) for _ in range(10)]
y = [random.randint(0, 50) for _ in range(10)]
colors = ["red","green","blue","yellow","pink","orange","purple","brown","cyan","magenta"]
plt.scatter(x, y, c=colors)
plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")
plt.show()
```

```python
import matplotlib.pyplot as plt
years = [2017, 2018, 2019, 2020, 2021, 2022]
bears = [8, 54, 93, 116, 137, 184]
dolphins = [150, 77, 32, 11, 6, 1]
whales = [80, 54, 100, 76, 93, 72]
labels = ["Bears", "Dolphins", "Whales"]
plt.figure(figsize=(12, 8))
for i in range(len(years)):
    data = [bears[i], dolphins[i], whales[i]]
    plt.subplot(2, 3, i+1)   # 2 rows, 3 columns
    plt.pie(data, labels=labels, autopct='%1.1f%%')
    plt.title(str(years[i]))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
# data
x = ["Red", "Silver", "Blue", "White", "Green"]
y = [123, 230, 78, 193, 12]
# a
plt.figure()
plt.bar(x, y)
plt.xlabel("Car Color")
plt.ylabel("Number Seen")
plt.title("Bar Chart - Default")
plt.show()
# b
plt.figure()
plt.bar(x, y, width=0.3)
plt.title("Bar Chart - Width 0.3 (Thinner Bars)")
plt.show()
# c
plt.figure()
plt.barh(x, y)
plt.title("Horizontal Bar Chart")
plt.show()
# d
colors = ["red", "silver", "blue", "white", "green"]
plt.figure()
plt.bar(x, y, color=colors)
plt.title("Bar Chart with Matching Colors")
plt.show()
# e
plt.figure()
plt.pie(y, labels=x, autopct='%1.1f%%')
plt.title("Car Color Distribution")
plt.show()
# f
myexplode = [0.3, 0, 0, 0, 0]
plt.figure()
plt.pie(y, labels=x, explode=myexplode, autopct='%1.1f%%')
plt.title("Pie Chart with Exploded Slice")
plt.show()
# pie chart(explode+shadow)
plt.figure()
plt.pie(y, labels=x, explode=myexplode, shadow=True, autopct='%1.1f%%')
plt.title("Pie Chart with Shadow")
plt.show()
# pie chart with legend
plt.figure()
plt.pie(y, autopct='%1.1f%%')
plt.legend(x, title="Car Colours:")
plt.title("Pie Chart with Legend")
plt.show()
#  pie chart with custom colors
mycolours = ["magenta", "green", "cyan", "red", "yellow"]
plt.figure()
plt.pie(y, labels=x, colors=mycolours, autopct='%1.1f%%')
plt.title("Pie Chart with Custom Colors")
plt.show()
```
